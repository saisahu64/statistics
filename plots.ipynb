```python
import matplotlib.pyplot as plt
```

## Pie Plot

```python
plt.pie([15, 25, 40, 60], labels=["A", "B", "C", "D"])
plt.show()
```

## Dot plot

```python
plt.scatter(["A", "B", "C", "D"], [5, 8, 20, 2])
plt.show()
```

## Bar Plot

```python
plt.bar(["A", "B", "C", "D"], [5, 8, 20, 2])
plt.show()
```

## Histogram

```python
plt.hist([1, 2, 2, 3, 3, 3, 4, 4, 5, 6, 7, 8, 8, 9, 10, 10, 10, 11, 12, 13, 14], bins=5, density=True)
plt.show()
```

## Box and Whishers plot

```python
plt.boxplot([[1, 2, 2, 3, 3, 3, 4, 4, 5, 6, 7, 8, 8, 9, 10, 10, 10, 11, 12, 13, 14]])
plt.plot()
```

## Scatter Plot

```python
plt.scatter([1, 2, 3, 4, 5], [1, 3, 4, 5, 6])
plt.show()
```
